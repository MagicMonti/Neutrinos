```python
from sympy.physics.matrices import mgamma
```

```python
def commutator(i,j):
    return mgamma(i)@mgamma(j) - mgamma(j)@mgamma(i)

def sigma(i,j):
    psi = Matrix([[1,0,0,0]])
    #psi = Matrix([[0,1,0,0]])
    return psi@mgamma(0)@commutator(i,j)@psi.T
```

```python
s = zeros(4)


for i in range(4):
    for j in range(4):
        s[i,j] = sigma(i,j)/4

s
```

```python
print_latex(s)
```
